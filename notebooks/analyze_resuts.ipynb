```python
cols = ['deepseek-chat', 'gemini-2.0-flash',
       'gemma-3-27b-it', 'gpt-4o-2024-08-06', 'gpt-4o-mini-2024-07-18',
       'llama-4-maverick-17b-128e-instruct']
grouped_df = output_sum_df.groupby(["Setting", "Dataset", "Model"])["Length Difference"].sum()
unstacked_df = grouped_df.unstack("Model").fillna("-")
unstacked_df[cols]
```

### AA

```python
import pandas as pd


aa_g_df = pd.read_csv("results/AA_generated_results.csv")
aa_g_df = aa_g_df[aa_g_df.Setting != "Setting6"]
aa_o_df = pd.read_csv("results/AA_original_results.csv")
aa_o_df["Source"] = "Original"
out = []

for i in range(1, 6):
    aa_o_df_copy = aa_o_df_copy.copy()
    aa_o_df_copy["Setting"] = f"Setting{i}"
    out.append(aa_o_df_copy)

aa_o_df = pd.concat(out, ignore_index=True)

aa_g_df.rename(columns={"LLM_Model": "Source"}, inplace=True)
aa_o_df.columns, aa_g_df.columns
```

```python
aa_cat_df = pd.concat([aa_o_df, aa_g_df], axis=0)
aa_cat_df
```

```python
a = aa_cat_df.groupby(["AA_Model", "Setting", "Dataset", "Source"])["Top_1"].sum().unstack("Source")
a_non_original = a[a.columns.to_list()[1:]]
a["Avg LLM"] = a_non_original.mean(axis=1).round(2).fillna("-")
a
```

```python
a = aa_cat_df.groupby(["AA_Model", "Setting", "Dataset", "Source"])["Top_3"].sum().unstack("Source")
a_non_original = a[a.columns.to_list()[1:]]
a["Avg LLM"] = a_non_original.mean(axis=1).round(2).fillna("-")
a
```

```python
a = aa_cat_df.groupby(["AA_Model", "Setting", "Dataset", "Source"])["Top_5"].sum().unstack("Source")
a_non_original = a[a.columns.to_list()[1:]]
a["Avg LLM"] = a_non_original.mean(axis=1).round(2).fillna("-")
a
```

```python
a = aa_cat_df.groupby(["AA_Model", "Setting", "Dataset", "Source"])["Top_10"].sum().unstack("Source")
a_non_original = a[a.columns.to_list()[1:]]
a["Avg LLM"] = a_non_original.mean(axis=1).round(2).fillna("-")
a
```

```python
av_o_df = pd.read_csv("results/AV_original_results.csv")
av_g_df = pd.read_csv("results/AV_generated_results.csv")
av_g_df = av_g_df[av_g_df.Setting != "Setting6"]

av_o_df["Source"] = "Original"
av_o_df_copy = av_o_df.copy()
out = []

for i in range(1, 6):
    av_o_df_copy = av_o_df_copy.copy()
    av_o_df_copy["Setting"] = f"Setting{i}"
    out.append(av_o_df_copy)
    
av_o_df = pd.concat(out, ignore_index=True)
av_o_df.rename(columns={"AA_Model": "AV_Model"}, inplace=True)

av_g_df.rename(columns={"LLM_Model": "Source", "AA_Model": "AV_Model"}, inplace=True)
av_cat_df = pd.concat([av_o_df, av_g_df], axis=0)
av_cat_df = av_cat_df[av_cat_df["AV_Model"] != "bert-base-uncased"].reset_index(drop=True)
```
