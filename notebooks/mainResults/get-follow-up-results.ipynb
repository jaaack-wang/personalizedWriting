### Get Setting 1 subsets

```python
def align_df1_to_df2(df1, df2, col1, col2, col3):
    """
    Aligns df1 to df2 based on matching values in columns `col1,` `col2` and `col3`,
    handling duplicates in both dataframes.

    Returns a new DataFrame with the same number of rows and order as df2,
    by selecting matching rows from df1.
    """
    df1_temp = df1.copy()
    df1_temp['_used'] = False
    matched_rows = []
    indices = []

    for _, row in df2.iterrows():
        match = df1_temp[
            (df1_temp[col1] == row[col1]) &
            (df1_temp[col2] == row[col2]) &
            (df1_temp[col3] == row[col3]) &
            (~df1_temp['_used'])
        ]

        if match.empty:
            return None

        first_match_idx = match.index[0]
        matched_rows.append(df1_temp.loc[first_match_idx])
        df1_temp.at[first_match_idx, '_used'] = True
        indices.append(first_match_idx)

    result = pd.DataFrame(matched_rows).drop(columns=['_used']).reset_index(drop=True)
    return indices
```

```python
import os

datasets = ["blog", "enron", "reddit", "CCAT50"]
models = ["gemini-2.0-flash", "gpt-4o-2024-08-06", "llama-4-maverick-17b-128e-instruct"]

for dataset in datasets:
    follow_eval_fp = f"dataset_followup/{dataset}_test.csv"
    seting1_prompt_fp = f"LLM_writing/Setting1/{dataset}/prompts.csv"
    setting4_prompt_fp = f"LLM_writing/Setting4/{dataset}/prompts.csv"

    follow_eval_df = pd.read_csv(follow_eval_fp)
    seting1_prompt_df = pd.read_csv(seting1_prompt_fp)
    setting4_prompt_df = pd.read_csv(setting4_prompt_fp)

    indices = align_df1_to_df2(seting1_prompt_df, follow_eval_df, 
                               "text", "author", "summary")
    assert follow_eval_df[["text", "author", "summary"]].equals(
        seting1_prompt_df.loc[indices][["text", "author", "summary"]].reset_index(drop=True)
    ), f"Mismatch in {dataset} dataset between follow_eval_df and seting1_prompt_df"
    assert follow_eval_df[["text", "author", "summary"]].equals(
        setting4_prompt_df.loc[indices][["text", "author", "summary"]].reset_index(drop=True)
    ), f"Mismatch in {dataset} dataset between follow_eval_df and setting4_prompt_df"

    os.makedirs(f"LLM_writing/Setting1_sub/{dataset}/", exist_ok=True)
    os.makedirs(f"LLM_writing/Setting4_sub/{dataset}/", exist_ok=True)
    seting1_prompt_df_sub = seting1_prompt_df.loc[indices].reset_index(drop=True)
    seting1_prompt_df_sub["OriginalIndex"] = indices
    seting1_prompt_df_sub.to_csv(f"LLM_writing/Setting1_sub/{dataset}/prompts.csv", index=False)
    setting4_prompt_df_sub = setting4_prompt_df.loc[indices].reset_index(drop=True)
    setting4_prompt_df_sub["OriginalIndex"] = indices
    setting4_prompt_df_sub.to_csv(f"LLM_writing/Setting4_sub/{dataset}/prompts.csv", index=False)

    for setting in [1,4]:
        for model in models:
            llm_fp = f"LLM_writing/Setting{setting}/{dataset}/{model}.csv"
            
            if not os.path.exists(llm_fp):
                print(f"File {llm_fp} does not exist.")
                continue                
            
            llm_df = pd.read_csv(llm_fp)
            llm_df_sub = llm_df.loc[indices].reset_index(drop=True)
            llm_df_sub["OriginalIndex"] = indices
            llm_df_sub.to_csv(f"LLM_writing/Setting{setting}_sub/{dataset}/{model}.csv", index=False)
            print(f"Saved {llm_fp} to LLM_writing/Setting{setting}_sub/{dataset}/{model}.csv")
```
